```python
# Kod för att ta hem metadatat på fordonsslag

# Fordon metadata

import os
import requests
import pandas as pd

# Kontrollera att 'save_directory' är definierad
try:
    save_directory
except NameError:
    raise NameError("Variabeln 'save_directory' är inte definierad. Var god definiera den i en separat cell.")

# Ange filnamnet
file_name = 'fordonsnr_kod_namn.csv'

# Bygg den fullständiga sökvägen
save_path = os.path.join(save_directory, file_name)

# URL för att hämta metadata om fordon
url = 'https://api.scb.se/OV0104/v1/doris/sv/ssd/START/TK/TK1001/TK1001A/FordonTrafik'

# Skicka GET-förfrågan för att hämta metadata
response = requests.get(url)

# Kontrollera om förfrågan lyckades
if response.status_code == 200:
    metadata = response.json()

    # Hitta regionvariabelns metadata
    for variable in metadata['variables']:
        if variable['code'] == 'Fordonsslag':
            fordon_codes = variable['values']
            fordon_names = variable['valueTexts']

            # Skapa en DataFrame med kod och namn
            df = pd.DataFrame({
                'kod': fordon_codes,
                'namn': fordon_names
            })

    # Spara DataFrame som CSV
    df.to_csv(save_path, index=False, encoding='utf-8-sig')  # UTF-8 med BOM för korrekt hantering av svenska tecken
    print(f"Data sparad som CSV på: {save_path}")
else:
    print(f"Misslyckades att hämta data: {response.status_code}")
```

```python
# Ändrar granulariteten för fordon_i_trafik_per_kommun istället för att vara års information printat på december (12:e månaden) så delat vi på 12 och 
# skapar genomsnittet av fordon i trafik över året (för varje månad alltså)

import pandas as pd

# Ladda CSV-filen från den angivna katalogen
file_path = save_directory + r'\fordon_i_trafik_per_kommun.csv'
df = pd.read_csv(file_path, dtype={'region': str, 'fordonsnr': str})  # Läser in 'region' och 'fordonsnr' som textvärden

# Skapa en lista över månader för varje år
def skapa_månader(år):
    return [f'{år}M{str(månad).zfill(2)}' for månad in range(1, 13)]

# Skapa en tom lista för nya rader
new_rows = []

# För varje rad i df, skapa 12 nya rader med samma 'region' och 'fordonsnr', men med månad från januari till december
for index, row in df.iterrows():
    # Hämta år från 'DatumID'
    år = row['DatumID'][:4]
    månader = skapa_månader(år)
    
    # Dela värdet på 12 och skapa en ny rad för varje månad
    månatligt_värde = row['Fordon i trafik'] / 12
    
    for månad in månader:
        new_row = row.copy()
        new_row['DatumID'] = månad  # Uppdatera DatumID med den nya månaden
        new_row['Fordon i trafik'] = månatligt_värde  # Sätt månatligt värde
        new_rows.append(new_row)

# Skapa en ny DataFrame med de nya raderna
new_df = pd.DataFrame(new_rows)

# Spara den uppdaterade DataFrame till en ny CSV-fil i samma katalog
new_file_path = save_directory + r'\uppdaterad_fordon_i_trafik_per_kommun.csv'
new_df.to_csv(new_file_path, index=False)

print("Filen har sparats till:", new_file_path)
```
